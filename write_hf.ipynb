```python
import os
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

id2label = {0: 'Доставка', 1: 'Магазин', 2: 'Товар'}

# Путь к модели
MODEL_PATH = "./final_model" 
#MODEL_PATH = "./models/kfold/fold_5"

tokenizer = AutoTokenizer.from_pretrained('cointegrated/rubert-tiny2')

if torch.cuda.is_available():
    device = 'cuda'
elif torch.backends.mps.is_available():
    device = 'mps' # я на маке сидел с M2
else:
    device = 'cpu'
device = torch.device(device)


model = AutoModelForSequenceClassification.from_pretrained(MODEL_PATH)
model.to(device)
model.eval()
```

```python
# 1. Устанавливаем и импортируем нужную утилиту (если еще не установлена)
# Сначала в терминале: pip install huggingface_hub
from huggingface_hub import login
import os

# 2. Авторизуемся по токену (создайте токен с правами "Write" на hf.co/settings/tokens)
HF_TOKEN = os.getenv('HF_WRITE_TOKEN')
login(token=HF_TOKEN)

# 3. Указываем название вашего будущего репозитория на HF
# Формат: "ваш_логин_на_hf/название_модели"
REPO_ID = "jorapro/my-rubert-tiny2-review-classifier"

# 4. Загружаем модель и токенизатор на хаб
print("Загрузка модели на Hugging Face...")
model.push_to_hub(REPO_ID)
tokenizer.push_to_hub(REPO_ID)

print(f"Готово! Модель доступна по ссылке: https://huggingface.co{REPO_ID}")
```
